```python
#Load dataset
import pandas as pd
```

```python
#Read excel file
df = pd.read_excel('electric_power_data.xlsx')
#print(df.head())
#print(df.columns)
print(df.info())
```

```python
import matplotlib.pyplot as plt
df['datetime'] = pd.to_datetime(df['datetime'])

#datetime to index
df.set_index('datetime', inplace = True)

plt.figure(figsize=(12, 8))
```

```python
#1
plt.subplot(2, 2, 1)
plt.plot(df.index, df['P(kW)'], color='red')
plt.title('Power Output Over Time')
plt.xlabel('Date')
plt.ylabel('Power Output(kW)')
plt.grid(True)

#2
plt.subplot(2, 2, 2)
plt.plot(df.index, df['Ua(V)'], color='green')
plt.title('Voltage Over Time')
plt.xlabel('Date')
plt.ylabel('Voltage(V)')
plt.grid(True)

#3
plt.subplot(2, 2, 3)
plt.plot(df.index, df['TEMP(°C)'], color='blue')
plt.title('Temperature Over Time')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.grid(True)

#4
plt.subplot(2, 2, 4)
plt.plot(df.index, df['H(%)'], color='orange')
plt.title('Humidity Over Time')
plt.xlabel('Date')
plt.ylabel('Humidity H(%)')
plt.grid(True)

plt.tight_layout()
plt.show()
```
